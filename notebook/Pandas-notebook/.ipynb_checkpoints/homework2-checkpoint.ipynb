```python
d3_1 = df.set_index(['department', 'job_title', 'gender']).reorder_levels([2,1,0])
d3_1.head()
```

```python
d3_2 = d3_1.reset_index('job_title')
d3_2.head()
```

```python
d3_3 = d3_2.rename_axis(index = {'gender': 'Gender'})
d3_3.head()
```

```python
new_index = d3_3.index.map(lambda x:(x[0]+'-'+x[1]))
d3_4 = d3_3.copy()
d3_4.index = new_index
d3_4.head()
```

```python
new_index2 = d4_3.index.map(lambda x:tuple(x.split('-')))
d3_5 = d3_4.copy()
d3_5.index = new_index2
d3_5.head()
```

```python
d3_6 = d3_5.rename_axis(index=['gender', 'department'])
d3_6.head()
```

```python
d3_7 = d3_6.reset_index().reindex(df.columns,axis=1)
d3_7.head()
```
